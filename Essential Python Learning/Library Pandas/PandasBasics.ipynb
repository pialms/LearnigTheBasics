# Pandas
Pandas is a powerful, open-source Python library used primarily for data manipulation, analysis, and cleaning. Built on top of NumPy, it is a fundamental tool for data scientists and analysts because it makes working with structured data (like spreadsheets or SQL tables) fast and intuitive.

## Basic Data Structure in Pandas

pandas provides two types of classes for handling data:

1. `Series`: a one-dimensional labeled array holding data of any type such as integers, strings, Python objects etc.

2. `DataFrame`: a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.

## What is Index

- In pandas, an Index is an immutable sequence used to store axis labels for Series and DataFrames. 
- It acts like an "address" for data, allowing it to identify, align, and retrieve specific rows or columns efficiently.

### Core Concepts
- **Purpose**: The pandas Index enables label-based access and ensures data stays aligned during operations (like adding two DataFrames with different row orders).
- **Immutability**: Once created, the values within an Index cannot be changed individually, though the entire Index can be replaced.
- **Default Index**: When we create a DataFrame without specifying one, pandas assigns a RangeIndex—an memory-efficient integer-based index starting at 0.

### Common Operations
- **Set Index**: Use `df.set_index('col1')` to turn an existing column into the row index.
- **Reset Index**: Use `df.reset_index()` to move the current index back into a column and return to a default integer index.
- **Check Uniqueness**: Use `df.index.is_unique` to verify if all labels are distinct.
- **MultiIndex**: A hierarchical index that allows multiple levels of labels on a single axis.

### Key Differences

- ***Performance***: df.at is significantly faster than df.loc for scalar lookups because it has less overhead. If you are loop-setting or getting individual cells, use .at.
- ***Capability***: df.loc is much more versatile. It allows for slicing (e.g., df.loc[1:5, 'A':'C']) and filtering with boolean masks, which .at cannot do.
- ***Slicing Behavior***: Unlike standard Python slicing, df.loc is inclusive of the end label (e.g., df.loc['a':'c'] includes 'c').

## Function Passing Method:
1. `df/series.apply()`  : The pandas `.apply()` method is a versatile tool used to apply a function along an axis of a DataFrame or to every element of a Series. It is often used for data transformations that aren't natively supported by built-in pandas methods. While powerful, `.apply()` is essentially a loop in disguise. 

  ### How it Works:

  - On a Series: It runs a function on every single value in the column.
  - On a DataFrame: It runs a function across entire rows or columns.axis=0 (default): Applies the function to each column.axis=1: Applies the function to each row.

2. `df/series.map()`  : `.map()` is a powerful tool used to transform data by applying a function, dictionary, or another Series to every element in a Series or DataFrame.

### Key Usage Scenarios:

- **Using a Dictionary (Series only)**: This is the most common way to substitute values (e.g., mapping category IDs to names).
- **Using a Function**: You can pass a Python function or a lambda to transform every value (e.g., squaring numbers or changing text case).
- **Handling Missing Values**: You can use the na_action='ignore' parameter to prevent the mapping from affecting NaN values.
- **Series.map()**: Operates on a single column. It can take a dictionary, a Series, or a function.
- **DataFrame.map()**: Operates on every element across the entire table simultaneously. This replaced the now-deprecated .applymap() method.


3. `applymap()` : has been deprecated
4. `df/series.replace()`  : method is used to swap out values in a Pandas DataFrame or Series dynamically. It is highly flexible, allowing for single value replacements, dictionary-based lookups, or pattern matching via regular expressions. Designed for replacing entire cell values (exact matches). It works on numbers, strings, and the whole DataFrame.

## Difference Between series.apply() and Series.map():

While they can achieve similar results, `Series.map()` and `Series.apply()` are not the same. The primary difference lies in what they accept as input and how they handle complex operations.

### Key Differences

1. **Input Types**:

- `Series.map()` accepts a function, a dictionary, or another Series. This makes it ideal for direct value substitution (e.g., changing 'A' to 1 using a dictionary).
- `Series.apply()` only accepts a function. It cannot take a dictionary or Series for direct mapping.

2. **Handling Arguments**:

- `Series.apply()` can take additional positional or keyword arguments (args and **kwargs) to pass to the function.
- `Series.map()` generally does not support passing extra arguments to the mapping function.

3. **NaN Handling**:

- `Series.map()` has a specific na_action parameter that allows you to ignore NaN values, preventing the function from being applied to them.
- `Series.apply()` does not have this specific parameter; it will apply the function to every element regardless of whether it is a null value.

4. **Performance**:

- `Series.map()` is often faster for simple element-wise transformations or dictionary-based replacements.
- `Series.apply()` is better suited for more complex logic that might involve conditions or external arguments.
